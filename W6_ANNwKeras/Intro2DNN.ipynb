#**MY FIRST DEEP NEURAL NETWORK (DNN)**

**Single layer feedforward network**

[2 inputs] -> [2 outputs]

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
```

```python
Perceptron = Sequential()
Perceptron.add(Dense(2, input_dim=2, activation="sigmoid"))
```

```python
print(Perceptron.summary())
```

```python
from keras.utils import plot_model
```

```python
plot_model(Perceptron, show_shapes=True)
```

**Deep Neural Network with 1 hidden layer**

```python
DNN1 = Sequential()
DNN1.add(Dense(15, input_dim=10,activation='relu'))
DNN1.add(Dense(2, activation='sigmoid'))
```

```python
print(DNN1.summary())
```

```python
plot_model(DNN1, show_shapes=True)
```

**Deep Neural Network with mulitple hidden layers**

```python
DNN2 = Sequential()
DNN2.add(Dense(15,input_dim=10, activation='relu'))
DNN2.add(Dense(10,activation='relu'))
DNN2.add(Dense(2,activation='sigmoid'))
```

```python
print(DNN2.summary())
```

```python
plot_model(DNN2, show_shapes=True)
```
