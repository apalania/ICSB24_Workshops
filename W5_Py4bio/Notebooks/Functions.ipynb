#**FUNCTIONS**


- Positional arguments should be defined before any named arguments, and in order

- Specifying named arguments in a function call is optional; unspecified named arguments will take their default values.


define a function with no arguments and no return values

```python
def print_text():
    print('this is text')
```

 call the function

```python
print_text()
```

define a function with one argument and no return values

```python
def print_this(x):
    print(x)
```

call the function - print_this

```python
print_this(3)                                                       # prints 3
n = print_this(3)                                                   # prints 3, but doesn't assign 3 to n because the function has no return statement
```

define a function with one argument and one return value

```python
def square_this(x):
    return x**2
```

include an optional docstring to describe the effect of a function

```python
def square_this(x):
    """Return the square of a number."""
    return x**2
```

call the function

```python
square_this(3)                                                      # prints 9
var = square_this(3)                                                # assigns 9 to var, but does not print 9
```

Define a function with two 'positional arguments' (no default values) and one 'keyword argument' (has a default value)

```python
def calc(a, b, op='add'):
    if op == 'add':
        return a + b
    elif op == 'sub':
        return a - b
    else:
        print('valid operations are add and sub')
```

call the function

```python
calc(10, 4, op='add')                                               # returns 14
calc(10, 4, 'add')                                                  # also returns 14: unnamed arguments are inferred by position
calc(10, 4)                                                         # also returns 14: default for 'op' is 'add'
calc(10, 4, 'sub')                                                  # returns 6
calc(10, 4, 'div')                                                  # prints 'valid operations are add and sub'
```

Return two values from a single function

```python
def min_max(nums):
    return min(nums), max(nums)
```

Return values can be assigned to a single variable as a tuple

```python
nums = [1, 2, 3]
min_max_num = min_max(nums)                                         # min_max_num = (1, 3)
```

return values can be assigned into multiple variables using tuple unpacking

```python
min_num, max_num = min_max(nums)                                    # min_num = 1, max_num = 3
```

**CALCULATING THE MOLECULAR WEIGHT OF A PROTEIN USING FUNCTION**

```python
def calculate_molecular_weight(prot_seq):
    mol_wt = {'A': 89.1, 'V': 117.1, 'L': 131.2, 'I': 131.2, 'P': 115.1,
        'F': 165.2, 'W': 204.2, 'M': 149.2, 'G': 75.1, 'S': 105.1,
        'C': 121.2, 'T': 119.1, 'Y': 181.2, 'N': 132.1, 'Q': 146.2,
        'D': 133.1, 'E': 147.1, 'K': 146.2, 'R': 174.2, 'H': 155.2}

    total_weight = sum(mol_wt.get(prot.upper(), 0) for prot in prot_seq)
    total_weight -= (18 * (len(prot_seq) - 1))

    return total_weight

prot_seq = input('Enter your protein sequence: ')
net_weight = calculate_molecular_weight(prot_seq)

print('The net molecular weight is: {0}' .format(net_weight))
```

Line wise explanation for the above code:

*Line 1*    : Function "calculate_molecular_weight is created, that takes prot_seq as the input.

*Line 2 -5* : Dictionary of molecular weights for each amino acid.

*Line 7 *   : calculate the total weight.

*Line 8*    : Subtract water weight for peptide bonds

*line 10*   : return total weight

*Line 12*   : Input the protein sequence from the user

*Line 13*   : use the "calculate_molecular_weight" function to calculate the molecular weight of the user given protein sequence.

*Line 15*   : Print the net molecular weight
