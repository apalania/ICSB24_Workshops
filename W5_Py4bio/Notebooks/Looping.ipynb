#**LOOPS**
- Two types of loops: for and while.

**RANGE**

Range returns a list of integers (Python 2) or a sequence (Python 3)

    -range([start],stop,[step])

```python
range(0, 3)     # returns [0, 1, 2]: includes start value but excludes stop value
range(3)        # equivalent: default start value is 0
range(0, 5, 2)  # returns [0, 2, 4]: third argument is the step value
```

**FOR LOOP**
  
  -for iterVar in obj:

  <
  code chunk >
  
  ...

```python
fruits = ['apple', 'banana', 'cherry']
for i in range(len(fruits)):
    print(fruits[i].upper())
```

for loop - recommended style


```python
for fruit in fruits:
    print(fruit.upper())
```

iterate through two things at once (using tuple unpacking)

```python
family = {'dad':'homer', 'mom':'marge', 'size':6}
for key, value in family.items():
    print(key, value)
```

use enumerate if you need to access the index value within the loop

```python
for index, fruit in enumerate(fruits):
    print(index, fruit)
```

**FOR LOOP TO IDENTIFY THE MOLECULAR WEIGHT OF A PROTEIN**

let us try with the smallest human protein - insulin - chain A - GIVEQCCTSICSLYQLENYCN or chain B - FVNQHLCGSHLVEALYLVCGERGFFYTPKA.

```python
prot_seq = input('Enter your protein sequence:')
mol_wt = {'A':89.1, 'V':117.1, 'L':131.2, 'I':131.2, 'P':115.1,
                'F':165.2, 'W':204.2, 'M':149.2, 'G':75.1, 'S':105.1,
                'C':121.2, 'T':119.1, 'Y':181.2, 'N':132.1, 'Q':146.2,
                'D':133.1, 'E':147.1, 'K':146.2, 'R':174.2, 'H':155.2}
total_weight = 0
for prot in prot_seq:
  total_weight = total_weight + mol_wt.get(prot.upper(), 0)
total_weight = total_weight - (18 * (len(prot_seq) - 1))
print('The net molecular weight is: {0}'.format(total_weight))
```

**Line wise explanation of the above code:**

  *line 1* : user need to enter a protein sequence and stored in "prot_seq".

  *line 2-5* : A dictionary for the amino acid molecular weights is created as "mol_wt".
  *line 6* : The total weight of the protien is set to 0 initially.

  *line 7-8* : for loop is used in line 7 to iterate over each element (prot) in prot_seq.
  
  *line 9* : a correction factor of 18 is added due to the fact that each bond involves the loss of a water molecule (with molecular weight of 18).
  
  *line 10* : Prints out the net molecular weight.

**WHILE LOOP TO IDENTIFY THE MOLECULAR WEIGHT OF A PROTEIN**

let us try with the smallest human protein - insulin - chain A - GIVEQCCTSICSLYQLENYCN or chain B - FVNQHLCGSHLVEALYLVCGERGFFYTPKA.




```python
prot_seq = input('Enter your protein sequence:')
mol_wt = {'A': 89.1, 'V': 117.1, 'L': 131.2, 'I': 131.2, 'P': 115.1,
    'F': 165.2, 'W': 204.2, 'M': 149.2, 'G': 75.1, 'S': 105.1,
    'C': 121.2, 'T': 119.1, 'Y': 181.2, 'N': 132.1, 'Q': 146.2,
    'D': 133.1, 'E': 147.1, 'K': 146.2, 'R': 174.2, 'H': 155.2}
total_weight = 0
index = 0
while index < len(prot_seq):
    amino_acid = prot_seq[index].upper()  # Get the amino acid at the current index
    total_weight += mol_wt.get(amino_acid, 0)  # Add its molecular weight
    index += 1  # Move to the next amino acid
total_weight -= (18 * (len(prot_seq) - 1))
print('The net molecular weight is: {0}'.format(total_weight))
```

**line wise explanation of thw above code**

*Line 1*    : Input the protein sequence from the user

*Line 2-5*  : Dictionary of molecular weights for each amino acid

*Line 6-7*  : Initialize total weight and index

*Line 8-11* : Calculate total molecular weight using a while loop

*Line 12*   : a correction factor of 18 is added due to the fact that each bond involves the loss of a water molecule (with molecular weight of 18).

*Line 13*   : Print the net molecular weight
